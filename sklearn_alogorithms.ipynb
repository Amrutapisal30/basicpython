```python
import numpy as np
import pandas as pd
df=pd.read_csv("iris_dataset.csv")
df
```

```python
df.iloc[:,1:4]
```

```python
x=df.iloc[:,1:4].values
```

```python
y=df.iloc[:,4].values
```

```python
y  #it display taget --- string
```

```python
#To convert string to integer ---used LabelEncoder
from sklearn.preprocessing import LabelEncoder
labelencoder_y= LabelEncoder()
y=labelencoder_y.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
y_pred=logmodel.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
28/30  # (9+11+8)/sum of all elements
```

```python
from sklearn.neighbors import KNeighborsClassifier   #classification alogorithm---wo distances pe kam karta hai...kom=nsa point kis point se near hai
classifier_knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2) #5 nearest neigbors ko check karega,decide mejority
classifier_knn.fit(x_train,y_train)
```

```python
y_pred=classifier_knn.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
9+11+7/9+11+7+1+2
```

```python
from sklearn.naive_bayes import GaussianNB
classifier_nb=GaussianNB()
classifier_nb.fit(x_train,y_train)
```

```python
y_pred=classifier_nb.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
27/30
```

```python
from sklearn.svm import SVC  #support vector machine algorithm
classifier_svm_sigmoid=SVC(kernel='sigmoid')   #kernal-sigmoid,linear,rbf,polynomial
classifier_svm_sigmoid.fit(x_train,y_train)
```

```python
y_pred=classifier_svm_sigmoid.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
27/30
```
