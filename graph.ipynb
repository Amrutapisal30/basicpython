```python
from matplotlib import pyplot
import numpy as np
y=np.array([15,25,45,66])
mylables=["Apple","Banana","Cherry","Mango"]
pyplot.pie(y,labels=mylables)
pyplot.legend("Four color")
pyplot.show()
```

```python
from matplotlib import pyplot
import numpy as np
x=np.array(["john","blake","martin","tiger"])
y=np.array([56,78,34,58])
pyplot.bar(x,y,color=['red','blue','orange','yellow'])
pyplot.xlabel("Name")
pyplot.ylabel("Marks")
pyplot.title("Marksheet")
pyplot.legend("marks")
pyplot.show()

#create a bar chart x- mobile name and y axis- price
```

```python
from matplotlib import pyplot
import numpy as np
x=np.array([10,25,37,42,56])
y=np.array([100,200,300,400,500])
pyplot.plot(x,y,marker="v",mfc="red",mec="yellow",ms=20)  #marker="o","h","v",mfc-marker face color,mec-marker edge color,ms-marker size
pyplot.show()

```

```python
from matplotlib import pyplot
import numpy as np
x=np.array([170,250,370,422,546])
#pyplot.plot(x,ls="--")
pyplot.plot(x,ls="dotted",c="red",lw="3",marker='o',mfc="blue")  #ls-line style, dashed lw-line width
pyplot.show()
```
