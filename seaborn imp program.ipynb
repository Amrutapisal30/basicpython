```python
import seaborn as sns
sns.get_dataset_names()
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df1=sns.load_dataset('iris')
df1
```

```python
sns.scatterplot(x="age",y="fare",data=df)
```

```python
sns.scatterplot(x="age",y="fare",data=df,palette=["red","green","black"],hue="class",size="pclass")
```

```python
sns.histplot(df['age'],kde=False,bins=5)#kde --smoothness probablity function T/F shows line
```

```python
sns.displot(df['age'],bins=15,kde=True)
```

```python
sns.boxplot(x=df['pclass'],y=df['fare'],palette=['red','green'])
```

```python
sns.boxplot(x=df['pclass'],y=df['fare'],palette=['red','green'],hue=df['sex'])
```

```python
sns.violinplot(x=df['pclass'],y=df['fare'],palette=['red','green'],hue=df['sex'])
```

```python
sns.stripplot(x=df['pclass'],y=df['fare'],palette=['red','green'],hue=df['sex'])
```
