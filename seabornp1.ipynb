```python
import seaborn as sns
import pandas as pd
sns.get_dataset_names()
```

```python
df=sns.load_dataset("titanic")
df
```

```python
sns.scatterplot(x="age",y="fare",data=df)
```

```python
sns.scatterplot(x="age",y="fare",data=df,palette=["red","green","black"],hue="class",size="pclass")
```

```python
sns.histplot(df['age'],kde=True,bins=5)#kde- smoothness probability function T/F show lines
```

```python
sns.distplot(df['age'],bins=5,kde=True)
```

```python
sns.boxplot(x=df['pclass'],y=df['fare'],palette=['red','green'])
```

```python
sns.pairplot(df)
```
